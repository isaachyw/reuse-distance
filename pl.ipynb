```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
N=500
data=np.random.randn(N)
count,bins_count = np.histogram(data,bins=10)
```

```python
pdf = count / sum(count)
```

```python
cdf = np.cumsum(pdf)
```

```python
plt.plot(bins_count[1:],pdf,color='red',label="PDF")
plt.legend()
```
